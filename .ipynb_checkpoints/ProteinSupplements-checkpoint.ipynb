```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
titles_list = []
rattings_list = []
discounts_list = []
prices_list = []
```

```python
for i in range (1,50):
  url = 'https://www.flipkart.com/food-nutrition/health-nutrition/protein-supplements/pr?sid=7jv%2Cowt%2Cd9u&otracker=categorytree&otracker=nmenu_sub_Sports%2C+Books+%26+More_0_Protein+Supplements&otracker=nmenu_sub_Sports%2C+Books+%26+More_0_Protein+Supplements&page'+str(i)
  response = requests.get (url)
  soup = BeautifulSoup (response.text,'lxml')
  titles = soup.find_all ('a',class_='wjcEIp')
  for title in titles:
    titles_list.append (title.text.strip())
  
  rattings = soup.find_all ('span',class_='Wphh3N')
  for ratting in rattings:
    rattings_list.append (ratting.text.strip())
  
  discounts = soup.find_all ('div',class_="UkUFwK")
  for discount in discounts:
    discounts_list.append (discount.text.strip())
  
  prices = soup.find_all ('div',class_='Nx9bqj')
  for price in prices:
    prices_list.append (price.text.strip())

print(titles_list)
print(rattings_list)
print(discounts_list)
print(prices_list)

min_length = min(len(titles_list),
                len(rattings_list),
                len(discounts_list),
                len(prices_list))

titles_list = titles_list [:min_length]
rattings_list = rattings_list [:min_length]
discounts_list = discounts_list [:min_length]
prices_list = prices_list [:min_length]

df = pd.DataFrame ({'Titles':titles_list,
                   'Rattings':rattings_list,
                   'Discount':discounts_list,
                   'Price':prices_list})
print(df)
```
