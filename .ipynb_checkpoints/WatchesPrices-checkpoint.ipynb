```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

titles_list = []
descs_list = []
discounts_list = []
prices_list = []
deliverys_type_list = []
```

```python
for i in range (1,20):
  url = 'https://www.flipkart.com/search?q=watches&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page'+str(i)
  response = requests.get (url)
  soup = BeautifulSoup (response.text,'lxml')
  
  titles = soup.find_all ('div',class_='syl9yP')
  for title in titles:
    titles_list.append (title.text.strip())
  
  descs = soup.find_all ('a',class_='WKTcLC BwBZTg')
  for desc in descs:
    descs_list.append (desc.text.strip())
  
  discounts = soup.find_all ('div',class_='UkUFwK')
  for discount in discounts:
    discounts_list.append (discount.text.strip())
  
  prices = soup.find_all ('div',class_='Nx9bqj')
  for price in prices:
    prices_list.append (price.text.strip())
  
  deliverys_type = soup.find_all ('div',class_='yiggsN')
  for delivery_type in deliverys_type:
    deliverys_type_list.append (delivery_type.text.strip())


print(titles_list)
print(descs_list)
print(discounts_list)
print(prices_list)
print(deliverys_type_list)

min_length = min(len(titles_list),
                len(descs_list),
                len(discounts_list),
                len(prices_list),
                len(deliverys_type_list))

titles_list = titles_list [:min_length]
descs_list = descs_list [:min_length]
discounts_list = discounts_list [:min_length]
prices_list = prices_list [:min_length]
deliverys_type_list = deliverys_type_list [:min_length]

df = pd.DataFrame ({'Title':titles_list,
                   'Description':descs_list,
                   'Discount':discounts_list,
                   'Value':prices_list,
                   'Comments':deliverys_type_list
                   })
print(df)
```
