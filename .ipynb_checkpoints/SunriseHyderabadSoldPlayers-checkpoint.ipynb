```python
import requestshttps:
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.iplt20.com/auction'
page = requests.get (url)
print(page)
```

```python
soup = BeautifulSoup (page.text,'lxml')
```

```python
soup.find_all ('table')[5]
soup.find_all ('table',class_= 'ih-td-tab auction-tbl sold-players t3 ddtf-processed')
table = soup.find_all ('table')[5]
soup.find_all ('th')
chenai_titles = table.find_all ('th')
titles = []
for i in chenai_titles:
  title = i.text
  titles.append(title)
  
print(", ".join(titles))
```

```python
# Extract rows (<tr>) and data cells (<td>)
rows = table.find_all('tr')
data = []

for row in rows:
    cells = row.find_all('td')
    cells_text = [cell.text.strip() for cell in cells]
    if cells_text:  # Ignore empty rows
        data.append(cells_text)

# Create a DataFrame with the extracted data
df = pd.DataFrame(data, columns=titles)

df
```
