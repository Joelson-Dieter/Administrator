```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

titles_list = []
sizes_list = []
rattings_list = []
discounts_list = []
prices_list = []
```

```python
for i in range (1,150):
  url = 'https://www.flipkart.com/furniture/pr?sid=wwe&marketplace=FLIPKART&param=218&fm=neo%2Fmerchandising&iid=M_a874be14-e02f-4600-8eca-2bf89aeb35e8_1_VMDN1VE1AUK9_MC.DF4ISSCHXW05&otracker=hp_rich_navigation_8_1.navigationCard.RICH_NAVIGATION_Furniture_DF4ISSCHXW05&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_8_L0_view-all&cid=DF4ISSCHXW05&page'+str(i)
  response = requests.get (url)
  soup = BeautifulSoup (response.text,'lxml')
  
  titles = soup.find_all ('a',class_='wjcEIp')
  for title in titles:
    titles_list.append (title.text.strip())
  
  sizes = soup.find_all ('div',class_='NqpwHC')
  for size in sizes:
    sizes_list.append (size.text.strip())
  
  rattings = soup.find_all ('span',class_='Wphh3N')
  for ratting in rattings:
    rattings_list.append (ratting.text.strip())
  
  discounts = soup.find_all ('div',class_='UkUFwK')
  for discount in discounts:
    discounts_list.append (discount.text.strip())
  
  prices = soup.find_all ('div',class_='Nx9bqj')
  for price in prices:
    prices_list.append (price.text.strip())

print(titles_list)
print(sizes_list)
print(rattings_list)
print(discounts_list)
print(prices_list)

min_length = min(len(titles_list),
                len(sizes_list),
                len(rattings_list),
                len(discounts_list),
                len(prices_list)
                )

titles_list = titles_list [:min_length]
sizes_list = sizes_list [:min_length]
rattings_list = rattings_list [:min_length]
discounts_list = discounts_list [:min_length]
prices_list = prices_list [:min_length]

df = pd.DataFrame ({'Product title':titles_list,
                   'Product size':sizes_list,
                   'Rattings':rattings_list,
                   'Discounts allowed':discounts_list,
                   'Value':prices_list
                   })
print(df)
```
