```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://gadgetworld.co.ke/product-category/laptops/?orderby=popularity'
r = requests.get (url)
print(r)
```

```python
soup = BeautifulSoup (r.text,'lxml')
print(soup)
```

```python
desc = soup.find_all ('h2',class_='woocommerce-loop-product__title')
desc_list = []
for i in desc:
  des = i.text
  desc_list.append (des)

print(desc_list)
```

```python
print(len(desc))
```

```python
prices = soup.find_all ('span',class_='price')
price_list = []
for i in prices:
  price = i.text
  price_list.append (price)

print(price_list)
```

```python
print(len(prices))
```

```python
df = pd.DataFrame ({'Description':desc_list,'Prices':price_list})
print(df)
```
