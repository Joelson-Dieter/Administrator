```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.hwestauctions.com/schedule.v4.php'
page = requests.get (url)
print(page)
```

```python
soup = BeautifulSoup (page.text,'lxml')
soup.find_all ('table')[2]
soup.find_all ('table',class_="sorttable_sorted")
table = soup.find_all ('table')[2]
soup.find_all ('tr')
headers = table.find_all ('tr')
titles = []
for i in headers:
  title = i.text
  titles.append (title)
  
df = pd.DataFrame (columns = titles)





  
```

```python
df
```

```python
rows = table.find_all ('td')
for i in rows [1:]:
  data = i.find_all ('tr')
  rows = [tr.text for tr in data]
  l = len (df)
  df.loc [l] = rows
  print (df)
  
```
