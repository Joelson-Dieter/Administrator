```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.jumia.co.ke/phones-tablets/'
r = requests.get (url)
print(r)
```

```python
soup = BeautifulSoup (r.text,'lxml')
print(soup)
```

```python
data = soup.find_all (['div',class_:='name','prc','data-oprc'])
print(data)
```

```python
prices = soup.find_all ('div',class_='prc')
print(prices)
```

```python
print(len(prices))
```

```python
phone_price = []
for i in prices:
  price = i.text
  phone_price.append (price)

print(phone_price)
```
