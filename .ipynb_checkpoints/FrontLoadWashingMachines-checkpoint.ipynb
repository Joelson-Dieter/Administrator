```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

titles_list = []
speeds_list = []
rattings_list = []
connectivities_list = []
warranties_list = []
prices_list = []
discounts_list = []
```

```python
for i in range (1,15):
  url = 'https://www.flipkart.com/home-kitchen/home-appliances/washing-machines/fully-automatic-front-load~function/pr?sid=j9e%2Cabm%2C8qx&otracker=nmenu_sub_TVs+%26+Appliances_0_Fully+Automatic+Front+Load&page'+str(i)
  response = requests.get (url)
  
  soup = BeautifulSoup (response.text,'lxml')
  titles = soup.find_all ('div',class_="KzDlHZ")
  for title in titles:
    titles_list.append (title.text.strip())
  
  speeds = soup.find_all ('li',class_="J+igdf")
  for speed in speeds:
    speeds_list.append (speed.text.strip())
  
  rattings = soup.find_all ('li',class_='J+igdf')
  for ratting in rattings:
    rattings_list.append (ratting.text.strip())
  
  connectivities = soup.find_all ('li',class_='J+igdf')
  for connectivity in connectivities:
    connectivities_list.append (connectivity.text.strip())
  
  warranties = soup.find_all ('li',class_='J+igdf')
  for warranty in warranties:
    warranties_list.append(warranty.text.strip())
  
  prices = soup.find_all ('div',class_='Nx9bqj _4b5DiR')
  for price in prices:
    prices_list.append (price.text.strip())
  
  discounts = soup.find_all ('div',class_='UkUFwK')
  for discount in discounts:
    discounts_list.append (discount.text.strip())

print(titles_list)
print(speeds_list)
print(rattings_list)
print(connectivities_list)
print(warranties_list)
print(prices_list)
print(discounts_list)

min_length = min(len(titles_list),
                len(speeds_list),
                len(rattings_list),
                len(connectivities_list),
                len(warranties_list),
                len(prices_list),
                len(discounts_list)
                )

titles_list = titles_list [:min_length]
speeds_list = speeds_list [:min_length]
rattings_list = rattings_list [:min_length]
connectivities_list = connectivities_list [:min_length]
warranties_list = warranties_list [:min_length]
prices_list = prices_list [:min_length]
discounts_list = discounts_list [:min_length]

df=pd.DataFrame ({'Titles':titles_list,
                 'Speed':speeds_list,
                 'Rattings':rattings_list,
                 'Conectivity':connectivities_list,
                 'Warranties':warranties_list,
                 'Value':prices_list,
                 'Doscount':discounts_list})

print(df)
```
