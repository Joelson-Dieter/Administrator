```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_counties_of_Kenya_by_population'
response = requests.get (url)
print(response)
```

```python
soup = BeautifulSoup (response.text,'lxml')
print(soup.prettify()[:100])
```

```python
soup.find_all ('table')
soup.find ('table',class_='wikitable sortable jquery-tablesorter')
table = soup.find_all ('table')
soup.find_all ('th')
headers = soup.find_all ('th')
titles = []
for i in headers:
  title = i.text
  titles.append (title)
  
print(titles)
```

```python
df = pd.DataFrame (columns = titles)
df
```

```python
rows = soup.find_all ('tr')
for i in rows [1:]:
  data = i.find_all ('td')
  row = [tr.text.strip() for tr in data]
  l = len (df)
  df.loc [l] = row

df
```
