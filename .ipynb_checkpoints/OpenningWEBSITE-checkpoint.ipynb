```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
web = 'https://www.adamchoi.co.uk/overs/detailed'
path = r"C:\Users\Administrator\OneDrive\Documents\Data_Entry\PythonDOCS\chromedriver.exe"
service = Service (excutable_path=path)
driver = webdriver.Chrome (service=service)
driver.get (web)
all_matches_button = driver.find_element(by='xpath',value = '//label[@analytics-event="All matches"]')
all_matches_button.click()

from selenium.webdriver.common.by import By

# Find all rows in the table
matches = driver.find_elements(By.TAG_NAME, 'tr')

# Initialize empty lists to store data
dates_list = []
home_team = []
scores_list = []
away_team_list = []

# Iterate through each match row
for match in matches:
    try:
        # Extract date
        date = match.find_element(By.XPATH, './/td[1]').text
        dates_list.append(date)
        
        # Extract home team
        home = match.find_element(By.XPATH, './/td[2]').text
        home_team.append(home)
        
        # Extract score
        score = match.find_element(By.XPATH, './/td[3]').text
        scores_list.append(score)
        
        # Extract away team (corrected column index to `.//td[4]`)
        away_team = match.find_element(By.XPATH, './/td[4]').text
        away_team_list.append(away_team)
    except Exception as e:
        print(f"Error processing row: {match.get_attribute('outerHTML')}\n{e}")

# Print the results
print(dates_list)
print(home_team)
print(scores_list)
print(away_team_list)

    
    

    

```
